### 10. How Many People Have Heart Disease & How Many Don't Have Heart Disease ?

```python
data['target'].value_counts()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='target', data=data)
plt.title("Heart Disease Distribution")
plt.xlabel("Target (0 = No Disease, 1 = Disease)")
plt.ylabel("Count")
plt.show()
```

### 11. Which Gender Has The Most Heart Disease?

### 12. Find Gender Distribution According to The Target Variable.

```python
sns.countplot(x='sex',hue='target',data=data)
plt.xticks([1,0],['Male','Female'])
plt.legend(labels=['No-Disease','Disease'])
plt.show()
```

### 13. Check Age Distribution

```python
sns.histplot(data['age'], bins=20, kde=True)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.show()
```

### 14. Check Chest Pain Type

```python
### Chest pain type (4 values)
### Value 0: typical angina
### Value 1: atypical angina
### Value 2: non-anginal pain
### Value 3: asymptomatic 
```

```python
data['cp'] = data['cp'].map({
    0: "Typical Angina",
    1: "Atypical Angina",
    2: "Non-anginal Pain",
    3: "Asymptomatic"
})

plt.figure(figsize=(8,5))
sns.countplot(x='cp', data=data)
plt.xticks(rotation=30)
plt.title("Chest Pain Type Distribution")
plt.tight_layout()
plt.show()
```

### 15. Show The Chest Pain Distribution As Per Target Variable

```python
sns.countplot(x='cp',hue='target',data=data)
plt.legend(labels=['No-Disease','Diseases'])
plt.show()
```

### 16. Show Fasting Blood Sugar Distribution According To Target Variable

```python
sns.countplot(x='fbs',hue='target',data=data)
plt.legend(labels=['No-Disease','Diseases'])
plt.show()
```

### 17 Compare Resting Blood Pressure As Per the Gender

```python
g = sns.FacetGrid(data, hue='sex', aspect=4)
g.map(sns.kdeplot, 'trestbps', fill=True)
g.add_legend()
plt.show()
```
